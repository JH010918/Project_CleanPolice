```python
# 해상도 조절
from PIL import Image
import os

# ==========================
# 설정
# ==========================
input_folder = "room_before"  # 원본 이미지 폴더
output_folder = "room_after"  # 전처리 이미지 저장 폴더
target_size = (256, 256)      # 원하는 해상도

# 출력 폴더 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# ==========================
# 이미지 전처리
# ==========================
for filename in os.listdir(input_folder):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):  # 이미지 파일만 처리
        img_path = os.path.join(input_folder, filename)
        try:
            image = Image.open(img_path).convert("RGB")  # RGB 변환
            image_resized = image.resize(target_size, Image.Resampling.LANCZOS)
            save_path = os.path.join(output_folder, filename)
            image_resized.save(save_path)
            print(f"{filename} -> {target_size} 저장 완료")
        except Exception as e:
            print(f"{filename} 처리 실패: {e}")

print("모든 이미지 전처리 완료!")
```

```python
from ultralytics import YOLO
import os
import shutil

# ==========================
# 폴더 설정
# ==========================
INPUT_FOLDER = "room_before"          # 크롤링한 이미지
OUTPUT_FOLDER = "clean_dataset" # 사람 없는 이미지
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# ==========================
# YOLO 모델 로드
# ==========================
model = YOLO("yolov8n.pt")  # 가장 가벼운 모델

count = 0
removed = 0

# ==========================
# 이미지 검사
# ==========================
for file in os.listdir(INPUT_FOLDER):
    img_path = os.path.join(INPUT_FOLDER, file)
    if not file.lower().endswith((".jpg", ".png", ".jpeg")):
        continue
    results = model(img_path)
    person_detected = False
    for r in results:
        for cls in r.boxes.cls:
            if int(cls) == 0:   # 0 = person
                person_detected = True

    # ==========================
    # 사람 없는 이미지만 저장
    # ==========================
    if not person_detected:
        shutil.copy(img_path, os.path.join(OUTPUT_FOLDER, file))
        count += 1
        print(f"저장: {file}")
    else:
        removed += 1
        print(f"사람 포함 제거: {file}")

print("\n완료")
print("저장:", count)
print("제거:", removed)
```
